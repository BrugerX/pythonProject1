```python
import ExtractorsAndTables as eta
import Browser as brwsr
import pandas as pd
import utility.webscrapingUtil as wut
import LotData.LotData as ld

closed_LID = wut.getRandomClosedLID()
timestamp = wut.getTimeStamp()

latest_bid_table = eta.LatestBidTable(timestamp,brwsr.LotApi.getLotDescription(closed_LID))
bids_table = eta.BidsTable(timestamp,brwsr.BidApi.getBids(closed_LID))
metadata = eta.MetadataExtractor(123,123,111,123)

bids_df = bids_table.getDataframeCopy()
lbids_df = latest_bid_table.getDataframeCopy()
```

```python
bid_record = ld.BidRecord({"meta_data":metadata,"bid_data":bids_table,"latest_bid_data":latest_bid_table})
```
